```python
# connect to webdriver
options = webdriver.ChromeOptions()
options.add_argument("start-maximized")
driver = uc.Chrome(executable_path='chromedriver.exe', options=options)
```

```python
url = 'https://archive.ics.uci.edu/ml/index.php'
driver.get(url)
time.sleep(2)
```

```python
all_dataset_search =  driver.find_element_by_xpath("//span[@class='whitetext']/a/font")
all_dataset_search.click()
```

```python
dataset = []
data_type = []
data_type_list = []
task = []
attribute_type = []
no_of_instances = []
no_of_attribute = []
year = []

Dataset = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr//td[@valign='top'][2]/table[2]//tr//tr//td[2]//p")
Data_type = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr//td[@valign='top'][2]/table[@cellpadding='5']//tbody//tr//td[2]//p")
Task = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr//td[@valign='top'][2]/table[@cellpadding='5']//tbody//tr//td[3]//p")
Attribute_type = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr//td[@valign='top'][2]/table[@cellpadding='5']//tbody//tr//td[4]//p")
No_of_instances = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr//td[@valign='top'][2]/table[@cellpadding='5']//tbody//tr//td[5]//p")
No_of_attribute = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr//td[@valign='top'][2]/table[@cellpadding='5']//tbody//tr//td[6]//p")
Year = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr//td[@valign='top'][2]/table[@cellpadding='5']//tbody//tr//td[7]//p")

for i in Dataset:
    dataset.append(i.text.replace("\n", " "))
    
for i in Data_type:
    data_type.append(i.text.replace("\n", " "))
    
for i in range(0,len(data_type),2):
    data_type_list.append(data_type[i])

for i in Task:
    task.append(i.text.replace("\n", " "))

for i in Attribute_type:
    attribute_type.append(i.text.replace("\n", " "))
    
for i in No_of_instances:
    no_of_instances.append(i.text.replace("\n", " "))  
    
for i in No_of_attribute:
    no_of_attribute.append(i.text.replace("\n", " "))
    
for i in Year:
    year.append(i.text.replace("\n", " "))
    
data_type_list.pop(0)
task.pop(0)
attribute_type.pop(0)
no_of_instances.pop(0)
no_of_attribute.pop(0)
year.pop(0)
```

```python
Dataset_details=pd.DataFrame({}) 
Dataset_details['Name']=dataset  
Dataset_details['Data Type']=data_type_list
Dataset_details['Task']=task
Dataset_details['Attribute Type']=attribute_type
Dataset_details['No of Instances']=no_of_instances
Dataset_details['No of Attribute']=no_of_attribute
Dataset_details['Year']=year
Dataset_details
```
